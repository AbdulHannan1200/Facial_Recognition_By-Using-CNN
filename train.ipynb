```python
import cv2
import os
import numpy as np
from PIL import Image
import pickle

#BASE_DIR = os.path.dirname(os.path.abspath("attandence_openCV"))
#image_dir = os.path.join(BASE_DIR, "images")

BASE_DIR = os.path.dirname(os.path.abspath("Attendance_Facial_Recognition"))
image_dir = os.path.join(BASE_DIR, "images")
print(image_dir)
```

```python

face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_alt2.xml')
side_face_cascade = cv2.CascadeClassifier('data/haarcascade_profileface.xml')
#recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer = cv2.face.LBPHFaceRecognizer_create()

current_id = 0
label_ids = {}
y_labels = []
x_train = []
count=0
for root, dirs, files in os.walk(image_dir):
    for file in files:
        count=count+1
        if file.endswith("jpeg") or file.endswith("jpg")or file.endswith("png"):
            path = os.path.join(root, file)
            label = os.path.basename(root).replace(" ", "-").lower()
            print(label, path)
            print(count)
            if not label in label_ids:
                label_ids[label] = current_id
                current_id += 1
            id_ = label_ids[label]
            #print(label_ids)
            #y_labels.append(label) # some number
            #x_train.append(path) # verify this image, turn into a NUMPY arrray, GRAY
            pil_image = Image.open(path).convert("L") # grayscale
            size = (350, 350)
            final_image = pil_image.resize(size, Image.ANTIALIAS)
            image_array = np.array(final_image, "uint8")
            #print(image_array)
            faces = face_cascade.detectMultiScale(image_array, scaleFactor=1.5, minNeighbors=5)
            sides_faces=side_face_cascade.detectMultiScale(image_array, scaleFactor=1.5, minNeighbors=5)
            
            for (x,y,w,h) in faces:
                roi = image_array[y:y+h, x:x+w]
                x_train.append(roi)
                y_labels.append(id_)
     
            for (xx,yy,ww,hh) in sides_faces:
                    #cv2.rectangle(roi_color,(xx,yy),(xx+ww,yy+hh),(0,255,0),2)
                roi2 = image_array[y:yy+hh, x:xx+ww]
                x_train.append(roi2)
                y_labels.append(id_)



#print(y_labels)
#print(x_train)

#with open("face-labels.pickle", 'wb') as f:
 #   pickle.dump(label_ids, f)

#recognizer.train(x_train, np.array(y_labels))
#recognizer.save("face-trainner.yml")
```

```python
import pickle 
pickle_out=open("face-labels.pickle", 'wb')
pickle.dump(label_ids,pickle_out)
pickle_out.close()

pickle_out=open("X_train.pickle","wb")
pickle.dump(x_train,pickle_out)
pickle_out.close()

pickle_out=open("y.pickle","wb")
pickle.dump(y_labels,pickle_out)
pickle_out.close()
```

```python
recognizer.train(x_train, np.array(y_labels))
recognizer.save("face-trainner.yml")
```
